```python
def clean_tags(soup):
    """
    Code to decompose tags of a beautifulSoup object. In this case it will allow us to remove/forget the
    'sup' tags that have those references (ex: [1])
    """
    for tag in soup.find_all(['sup', 'span']):
        tag.decompose()
    

def get_info_box(url):
       
    resposta = requests.get(url)
    content = resposta.content    #Sacar o conteudo HTML da web page
    site = BeautifulSoup(content, 'html.parser')
    
    #once i got to the url ou a specific movie:
    info_table = site.find('table', attrs={'class': 'infobox vevent'})    #get the table we want
    info_rows = info_table.findAll('tr')                #and now the rows
    
    clean_tags(site)
    
    movie_info = {}
    content_value = None
    for index, row in enumerate(info_rows):
        if index == 0:
            movie_info['Title'] = row.find('th').text
        if index == 1:
            continue
        else:
            content_key = row.find('th').get_text() 
            try:
                content_value = row.find('td', attrs={'class': 'infobox-data'}).get_text().replace('\xa0', "")
                movie_info[content_key] = content_value
            except:
                pass

            if content_value == None:
                try:
                    content_value = row.find('td', attrs={'class': 'infobox-data'}).find('li').get_text().replace('\xa0', "")
                    movie_info[content_key] = content_value
                except:
                    pass

            if content_value == None:
                try:
                    content_value = row.find('td', attrs={'class': 'infobox-data'}).find('a').get_text().replace('\xa0', "")
                    movie_info[content_key] = content_value
                except:
                    pass
                
    lista = []
    for key, value in movie_info.items():
        lista.append([value])

    new = []
    for row in lista:
        for char in row:
            x = char.replace('\n', ', ')
            x = x.replace(', ,', ',')
            if x[:2] == ', ':
                x = x[2:]
            if x[-2:] == ', ':
                x = x[:-2]
            new.append(x)
            
    movie = {}
    for i, key in enumerate(movie_info.keys()):
        movie[key] = new[i]
        
    return movie

                
    
```

## Save/Reload Movie Data as a json file

```python
import json

def save_data(title, data):
    
    with open(title, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
def load_data(title):
    
    with open(title, encoding='utf-8') as f:
        return json.load(f)
```

```python
save_data('disney_data.json', movie_info_list)
```
